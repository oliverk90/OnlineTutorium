```python
%matplotlib inline
import requests
import numpy as np
import time
from datetime import datetime
from mpl_finance import candlestick2_ohlc
import matplotlib.pyplot as plt

#get data
url = "https://api.binance.com/api/v1/klines?symbol=BTCUSDT&interval=1m"
url = "https://api.binance.com//api/v3/klines?symbol=BTCUSDT&interval=1m"
url = "https://api.binance.com//api/v3/klines?symbol=ETHBTC&interval=1m"
dict_data = dict()
t1 = time.time()
response = requests.get(url)
data = response.json()

#transfer data to ohlc format
d = np.asarray(data)
d = d.astype(float)
da= d[:,0]
o= d[:,1]
h= d[:,2]
l= d[:,3]
c= d[:,4]

#plot scatter
ind = range(100)
plt.plot(h[ind],'b')
plt.plot(l[ind],'g')
ms = 10
da=[datetime.fromtimestamp(ddd/1000) for ddd in da]
plt.plot(h[ind],'b.',markersize=ms)
plt.plot(l[ind],'g.',markersize=ms)
plt.xlabel('timestep [#]')
plt.ylabel('BTC-price [USD]')
plt.legend(["highest price","lowest price"])
plt.title("Plot of BTC-price - 1 minute steps")
plt.show()


fig, ax = plt.subplots()
ind = range(100)
candlestick2_ohlc(ax, o[ind],h[ind],l[ind],c[ind], width=1, colorup='g')
plt.title("CandleStick Plot - 1 minute steps")
plt.xlabel('timestep [#]')
plt.ylabel('BTC-price [USD]')
```
